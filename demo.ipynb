```python
# Import network code
from nn import nn

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import imageio # for saving GIFs
```

### Test Case 1: $y = sin(2\pi x)$

```python
np.random.seed(0)
X = np.random.uniform(0, 1, size=(1, 100))
y = np.sin(2*np.pi*X)

network = nn([1, 10, 10, 1])
network.train(X, y, num_epochs=200)
preds = network.forward(X)

order = np.argsort(X[0])
X = X[:, order]
y = y[:, order]
preds = preds[:, order]

# Plotting the results
plt.plot(X[0], y[0], 'o', label='True Function')
plt.plot(X[0], preds[0], '-', label='Predicted Function')
plt.legend()
plt.title('Network Prediction vs True Function')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

### Test Case 2: $y=1\{x>0.5\}$

```python
np.random.seed(0)
X = np.random.uniform(0, 1, size=(1, 100))
y = X > 0.5

network = nn([1, 10, 10, 1])
network.train(X, y, num_epochs=200)
preds = network.forward(X)

order = np.argsort(X[0])
X = X[:, order]
y = y[:, order]
preds = preds[:, order]

# Plotting the results
plt.plot(X[0], y[0], 'o', label='True Function')
plt.plot(X[0], preds[0], '-', label='Predicted Function')
plt.legend()
plt.title('Network Prediction vs True Function')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

### Test Case 3: XOR Function

```python
np.random.seed(0)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]).T
y = np.array([[0], [1], [1], [0]]).T  # XOR function

network = nn([2, 10, 10, 1])
network.train(X, y, num_epochs=100)
preds = network.forward(X)

print("XOR Predictions:")
print(np.round(preds))
```

### Saving learning progress as a GIF

```python
np.random.seed(0)
X = np.random.uniform(0, 1, size=(1, 100))
y = np.sin(2*np.pi*X)
# y = X > 0.5  # Change to indicator function

network = nn([1, 10, 10, 1])
network.train(X, y, num_epochs=200, plot_progress=True)


frames = []
for epoch in range(0, 200, 10):
    image = imageio.v2.imread(f'./images/epoch_{epoch}.png')
    frames.append(image)

# Save GIF
imageio.mimsave('./images/sin_func.gif', frames, loop = 4, fps=5)
```
